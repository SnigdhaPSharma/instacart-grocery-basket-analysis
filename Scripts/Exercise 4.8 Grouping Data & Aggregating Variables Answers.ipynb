### Answer 4: Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
# Create max order column
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Check the dataframe for the new max_order column
ords_prods_merge.head(15)
```

```python
# Assign a flag based on the number of orders to each customer
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal Customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular Customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New Customer'
```

```python
# Check the fequency of different loyalty flags
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Check the relevant columns of the first few rows
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

### Answer 6: The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:

##### o	If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
##### o	If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”


```python
# Create mean price column
ords_prods_merge['mean_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Check the dataframe for the new mean_price column
ords_prods_merge.head(15)
```

```python
# Assign a flag based on the number of orders to each customer
ords_prods_merge.loc[ords_prods_merge['mean_price'] >= 10, 'spending_flag'] = 'High Spender'
ords_prods_merge.loc[ords_prods_merge['mean_price'] < 10, 'spending_flag'] = 'Low Spender'
```

```python
# Check the fequency of the two spending flags
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

### Answer 7:	In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
#### o	If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
#### o	If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
#### o	If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”


```python
# Create median_order_period column
ords_prods_merge['median_order_period'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Check the dataframe for the new median_order_period column
ords_prods_merge.head(60)
```

```python
# Assign a flag based on the frequency of ordering
ords_prods_merge.loc[ords_prods_merge['median_order_period'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['median_order_period'] <= 20) & (ords_prods_merge['median_order_period'] > 10), 'order_frequency_flag'] = 'Regular Customer'
ords_prods_merge.loc[ords_prods_merge['median_order_period'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
# Check the fequency of the frequency flags
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```
